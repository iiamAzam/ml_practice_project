```python
import pandas as pd 
from mpl_toolkits.mplot3d import axes3d
from matplotlib.animation import FuncAnimation


data = pd.read_csv("data.csv")

fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
x= data["Duration"]
y= data["Pulse"]
z= data["Calories"]
sc=ax.plot(x,y,z, color="red")

def rotate(angle):
    ax.view_init(elev=20, azim=angle)

ani = FuncAnimation(fig, rotate, frames=range(0, 360, 2), interval=100)


plt.show()
```
